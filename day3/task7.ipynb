```python
import numpy as np
import matplotlib.pyplot as plt

# 
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False  # 

# 
plt.rcParams['figure.dpi'] = 300

# x-551000
x = np.linspace(-5, 5, 1000)

# yx
y = x ** 3

# 
plt.figure(figsize=(8, 6))

# 
plt.plot(x, y, 'b-', linewidth=2, label=r'$y = x^3$')

# 
plt.grid(True, linestyle='--', alpha=0.7)

# 
plt.xlim(-5, 5)
plt.ylim(-125, 125)

# 
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.title('', fontsize=14)

# 
plt.legend(loc='upper left', fontsize=12)

# 
plt.tight_layout()
plt.show()
```

```python

# 
year = np.arange(2000, 2021).astype(np.str_)
month = np.random.randint(1, 13, size=20).astype(np.str_)
day = np.random.randint(1, 31, size=20).astype(np.str_)
date = np.array([])

for i in range(20):
    a = np.array([year[i], month[i], day[i]])
    b = ['/'.join(a)]  # 
    date = np.append(date, b)

# 
sales = np.random.randint(500, 2000, size=len(date))

# 
plt.xticks(range(0, len(date), 2), [':%s' % i for i in date][::2], rotation=45, color='red')
plt.plot(date, sales)
plt.show()
```
