```python
import numpy as np
import matplotlib.pyplot as plt

# 设置中文字体
plt.rcParams["font.family"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 设置图片清晰度
plt.rcParams['figure.dpi'] = 300

# 生成x值，范围从-5到5，共1000个点
x = np.linspace(-5, 5, 1000)

# 计算对应的y值（x的三次方）
y = x ** 3

# 创建图形和坐标轴
plt.figure(figsize=(8, 6))

# 绘制三次方函数曲线，设置为蓝色
plt.plot(x, y, 'b-', linewidth=2, label=r'$y = x^3$')

# 添加网格线
plt.grid(True, linestyle='--', alpha=0.7)

# 设置坐标轴范围
plt.xlim(-5, 5)
plt.ylim(-125, 125)

# 添加坐标轴标签和标题（中文显示）
plt.xlabel('x轴', fontsize=12)
plt.ylabel('y轴', fontsize=12)
plt.title('三次方函数图像', fontsize=14)

# 添加图例
plt.legend(loc='upper left', fontsize=12)

# 显示图像
plt.tight_layout()
plt.show()
```

```python

# 生成日期数据
year = np.arange(2000, 2021).astype(np.str_)
month = np.random.randint(1, 13, size=20).astype(np.str_)
day = np.random.randint(1, 31, size=20).astype(np.str_)
date = np.array([])

for i in range(20):
    a = np.array([year[i], month[i], day[i]])
    b = ['/'.join(a)]  # 组合年月日
    date = np.append(date, b)

# 生成随机销量数据
sales = np.random.randint(500, 2000, size=len(date))

# 绘制图形
plt.xticks(range(0, len(date), 2), ['日期:%s' % i for i in date][::2], rotation=45, color='red')
plt.plot(date, sales)
plt.show()
```
